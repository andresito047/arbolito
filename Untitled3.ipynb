```python
import matplotlib.pyplot as plt
import time

# Lista de nmeros afortunados menores o iguales a 1000
lucky_numbers = [4, 7, 44, 47, 74, 77, 444, 447, 474, 477, 744, 747, 774, 777]

# Funcin para verificar si un nmero es casi afortunado
def is_almost_lucky(n):
    for ln in lucky_numbers:
        if n % ln == 0:
            return True
    return False

# Valores de n para probar (del 1 al 1000)
n_values = list(range(1, 1001))
execution_times = []

for n in n_values:
    start_time = time.time()
    is_almost_lucky(n)
    end_time = time.time()
    execution_times.append(end_time - start_time)

# Graficar
plt.figure(figsize=(10, 5))
plt.plot(n_values, execution_times, color='blue')
plt.xlabel('Valor de n')
plt.ylabel('Tiempo de ejecucin (s)')
plt.title('Tiempo de ejecucin del algoritmo para distintos valores de n')
plt.grid(True)
plt.show()
```
